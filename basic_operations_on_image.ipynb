## Read Image
Read image with BGR.

```python
img = cv2.imread('/content/drive/MyDrive/images/strawberry.jpg');
plt.imshow(img);
```

Change the color range BGR to RGB.

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB);
plt.imshow(img_RGB);
```

Change the color range BGR to Grayscale.

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY);
plt.imshow(img_gray);
```

```python
plt.imshow(img_gray, cmap='gray');
```

## Resize Image

```python
# resize image to 32x32
img_resized = cv2.resize(img_RGB,(32,32));
plt.imshow(img_resized);
```

It looks like pixel art :)

```python
# resize image to 16x16
img_resized2 = cv2.resize(img_RGB,(16,16));
plt.imshow(img_resized2);
```

## Rotate Image

```python
# horizontally
img_flip3 = cv2.flip(img_RGB, 1);
plt.imshow(img_flip3);
```

```python
# vertically
img_flip2 = cv2.flip(img_RGB, 0);
plt.imshow(img_flip2);
```

```python
# both horizontally and vertically
img_flip1 = cv2.flip(img_RGB, -1);
plt.imshow(img_flip1);
```

## Access Pixel Values

Accessing the pixel values of 32x32 strawberry image.

```python
pixel_value = img_resized[15,15]
print(pixel_value)
```

Access only the value of red pixel. (RGB)

```python
red = img_resized[15,15,0]
print(red)
```

Modify pixel values.

```python
plt.imshow(img_resized);
```

```python
img_resized[15,15] = [0,0,0]
for i in range(4):
  img_resized[17,15+i] = [0,0,0]
img_resized[15,18] = [0,0,0]  
plt.imshow(img_resized);
```

Better ways of accessing and modifying pixel values.

```python
print(img_resized.item(15,15,0))
```

Changing the value of RED because of the image is RGB.

```python
img_resized.itemset((15,15,0),150)
print(img_resized.item(15,15,0))
```

## Split and Merge Channels

```python
b, g, r = cv2.split(img)
plt.imshow(img);
```

```python
img_merged = cv2.merge((r, g, b))
plt.imshow(img_merged);
```
